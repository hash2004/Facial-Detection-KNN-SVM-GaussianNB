```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/Q2/dataset.csv', header=None)
df.columns = [str(i) for i in range(1, len(df.columns)+1)]
df
```

```python
#Pre-process the dataset by normalizing each face image vector to unit length (i.e., dividing each vector by its magnitude).
#The magnitude of a vector is the square root of the sum of the squares of its elements.
#The magnitude of a vector v is denoted as ||v||.
#The normalized vector v' is calculated as v' = v / ||v||.
#The magnitude of a vector can be calculated using the numpy.linalg.norm() function.
#The normalized vector can be calculated by dividing the vector by its magnitude.
df_normalized = df.apply(lambda x: x / np.linalg.norm(x), axis=1)
df_normalized
```

```python
df = df_normalized
```

```python
labels = []
for i in range(1, 11):  # for each person numbered from 1 to 10
    labels += [i] * 170  # add the person number 170 times

df['Label'] = labels
df
```

```python
#convert all columns from string to float except label
df = df.astype(float)
df
```

```python
# test the data by checking person changes after every 170 rows
for i in range(0, len(df), 170):
    print(df['Label'][i])
```

```python
from sklearn.model_selection import train_test_split

# Assuming df is your original dataframe
X = df.iloc[:, :-1].values # Features (all columns except the last one)
y = df.iloc[:, -1].values  # Labels (last column)

# Perform a stratified split to ensure 20 rows per person end up in the test set
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=20/170, stratify=y, random_state=42
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import numpy as np

class KNN:
    def __init__(self, k=3, distance_metric='euclidean'):
        self.k = k
        self.distance_metric = distance_metric

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def distance(self, x1, x2):
        if self.distance_metric == 'euclidean':
            return np.sqrt(np.sum((x1 - x2) ** 2))
        elif self.distance_metric == 'cosine':
            return 1 - np.dot(x1, x2) / (np.linalg.norm(x1) * np.linalg.norm(x2))

    def predict(self, X):
        predicted_labels = [self._predict(x) for x in X]
        return np.array(predicted_labels)

    def _predict(self, x):
        distances = [self.distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]
        most_common = max(set(k_nearest_labels), key=k_nearest_labels.count)
        return most_common

    def classification_report(self, y_true, y_pred):
        unique_classes = np.unique(y_true)
        report = {}
        accuracy = np.sum(y_true == y_pred) / len(y_true)

        for cls in unique_classes:
            precision, recall, f1_score, support = self._calculate_metrics(y_true, y_pred, cls)
            report[cls] = {
                'precision': precision,
                'recall': recall,
                'f1_score': f1_score,
                'support': support
            }

        report['accuracy'] = accuracy

        return self._format_report(report)
    
    def accuracy_score(self, y_true, y_pred):
        return np.sum(y_true == y_pred) / len(y_true)
    
    def _calculate_metrics(self, y_true, y_pred, cls):
        true_positives = np.sum((y_pred == cls) & (y_true == cls))
        false_positives = np.sum((y_pred == cls) & (y_true != cls))
        false_negatives = np.sum((y_pred != cls) & (y_true == cls))

        precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
        recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0
        f1_score = 2 * precision * recall / (precision + recall) if precision + recall > 0 else 0
        support = np.sum(y_true == cls)

        return precision, recall, f1_score, support

    def _format_report(self, report):
        formatted_report = "Classification Report\n"
        formatted_report += "---------------------\n"
        for cls, metrics in report.items():
            if cls != 'accuracy':
                formatted_report += f"Class {cls}:\n"
                for metric, value in metrics.items():
                    formatted_report += f"  {metric}: {value:.2f}\n"
            else:
                formatted_report += f"Accuracy: {report['accuracy']:.2f}\n"

        return formatted_report
```

```python
knn = KNN(k=3, distance_metric='euclidean')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

report = knn.classification_report(y_test, y_pred)
print(report)
```

```python
k_values = range(2, 12)
accuracies = []
for k in k_values:
    knn = KNN(k=k, distance_metric='euclidean')
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o')
plt.title('KNN Accuracy for Different k Values')
plt.xlabel('k Value')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
k_values = range(2, 12)
accuracies = []
for k in k_values:
    knn = KNN(k=k, distance_metric='cosine')
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o')
plt.title('KNN Accuracy for Different k Values')
plt.xlabel('k Value')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```
